## Introduction

>### **Overview**
>To complete my Data Analysis project I am using Market Basket Optimisation Dataset. 

>This data set contains information about 7500 transaction collected from marketplace. It consist of 20 columns item. 

>#### **Question that can analysed from this data set**
> If the rules are min_support = 0.004, min_confidence = 0.2, min_lift = 3, min_length = 2, People who bought what item also bought what item? and what the value of support?

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Data Preprocessing
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
dataset.head()
```

```python
dataset.info()
```

## Training Apriori and Visualising the Result

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.004, min_confidence = 0.2, min_lift = 3, min_length = 2)

#we can change min_support, min_confidence, min_lift, and min_length.
```

```python
# Visualising the results
results = list(rules)
```

```python
results=pd.DataFrame(results)
results
```

```python
results.sort_values('support',ascending=False)
```
